```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib

matplotlib.rcParams.update({'font.size':18})
matplotlib.rcParams.update({'font.family':'serif'})
```

```python
# draw a schematic figure
plt.figure(figsize=(6,6))

# High-value, immediate ToO: 30min lead time, 2hr observation
plt.scatter([30./ 60. / 24.], [2.], alpha=0.5, s=900, c='RoyalBlue')
plt.text(.04,2, 'SNe/ToO', fontsize=12, color='RoyalBlue')

# high-to-med value, short lead-time ToO (e.g. Tabby's Star)
plt.plot([1,1], [1, 4], lw=5, c='Navy')
plt.text(1.1, 3, 'next-night', fontsize=12, color='Navy', rotation=90)

# "Standard" proposal programs w/ half-night allocations
plt.scatter([90], [6], alpha=0.5, s=100, c='red')
plt.text(15, 6, 'Standard', color='red', fontsize=12)

# possible new range of proposal allocations
plt.plot([90,90], [2,12], alpha=0.5, c='red')
plt.text(50,11, 'Flexible Proposals', color='red', fontsize=12, rotation=90)


# plus APOLLO
plt.scatter([90], [1], c='grey')
plt.text(20,1,'APOLLO', fontsize=10)


plt.ylabel('Duration Needed (Hours)')
plt.xlabel('Schedule Lead-Time')
plt.xscale('log')
plt.xlim(0.01, 110);
plt.ylim(0, 13);
plt.xticks([1./24., 6./24., 1., 90.], ('1h', '6h', '1d', '90d'));
plt.savefig('example.png', dpi=150, bbox_inches='tight', pad_inches=0.25)
```
